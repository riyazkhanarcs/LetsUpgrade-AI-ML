```
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```
#Mount Google Drive

from google.colab import drive
drive.mount('/content/drive')
```

```
#Import Dataset

dataset = pd.read_csv('/content/drive/My Drive/ML and AI/Assignments/Dataset/Attrition Dataset/general_data.csv')
dataset.head()
```

```
#Check for null values

dataset.isna().sum()
```

```
#Preprocess the data, Convert the String columns in to numbers

label_encoder = preprocessing.LabelEncoder()
dataset['Attrition'] = label_encoder.fit_transform(dataset['Attrition'])
dataset['BusinessTravel'] = label_encoder.fit_transform(dataset['BusinessTravel'])
dataset['Department'] = label_encoder.fit_transform(dataset['Department'])
dataset['EducationField'] = label_encoder.fit_transform(dataset['EducationField'])
dataset['Gender'] = label_encoder.fit_transform(dataset['Gender'])
dataset['JobRole'] = label_encoder.fit_transform(dataset['JobRole'])
dataset['MaritalStatus'] = label_encoder.fit_transform(dataset['MaritalStatus'])
```

```
#Remove unnecessary columns, nulls and duplicates

dataset1 = dataset.drop(['EmployeeCount', 'EmployeeID', 'Over18', 'StandardHours'], axis=1)
dataset2 = dataset1.dropna()
main_dataset = dataset2.drop_duplicates()
main_dataset.columns
```

```
#Identify significant columns by using Random Forest

features = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
            'Education', 'EducationField', 'Gender', 'JobLevel', 'JobRole',
            'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked',
            'PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears',
            'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion',
            'YearsWithCurrManager']
rf_model = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
rf_model.fit(X=main_dataset[features], y=main_dataset['Attrition'])
```

```
#Print the OOB accuracy value

print("OOB Accuracy is ", rf_model.oob_score_)
```

```
#Identify the important features

for feature, imp in zip (features, rf_model.feature_importances_):
  print(feature, imp)
```

```
#Take The Important features from the above output and start building the decision tree
#Important features are Age, MonthlyIncome and TotalWorkingYears

#Initialize Decision tree classifier

tree_model = tree.DecisionTreeClassifier(max_depth=8)
```

```
#Build Model

predictors = pd.DataFrame([main_dataset['Age'], main_dataset['MonthlyIncome'], main_dataset['TotalWorkingYears']]).T
tree_model.fit(X=predictors, y=main_dataset['Attrition'])
```

```
#Generate Dot file so that you can visually see the decision tree

with open("Dtree_Attrition.dot", 'w') as f:
  f = tree.export_graphviz(tree_model, feature_names=['Age','MonthlyIncome','TotalWorkingYears'], out_file=f)
```

```
#Check your model's score

tree_model.score(X=predictors, y=main_dataset['Attrition'])
```
